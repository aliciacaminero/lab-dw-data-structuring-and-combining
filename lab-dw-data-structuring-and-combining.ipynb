```python
import pandas as pd

url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
company1_df = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
company2_df = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
company3_df = pd.read_csv(url3)
```

```python
company1_df
```

```python
company2_df
```

```python
company3_df
```

```python
def first_clean(df):
    df.rename(columns= {'GENDER': 'Gender', 'ST': 'State'}, inplace=True)
    df.columns = df.columns.str.replace(' ', '_')
    df.columns = df.columns.str.lower()
    return df

c1_df = first_clean(company1_df)
c1_df['customer_lifetime_value'] = c1_df['customer_lifetime_value'].str.replace('%', '')
c2_df = first_clean(company2_df)
c2_df['customer_lifetime_value'] = c2_df['customer_lifetime_value'].str.replace('%', '')
c3_df = first_clean(company3_df)
```

```python
company_concat = pd.concat([c1_df, c2_df, c3_df])
company_concat
```

```python
company_concat.drop_duplicates(inplace=True)
company_concat
```

```python
company_concat.dropna(inplace=True)
company_concat
```

```python
#Borrar columna unnamed:_0, no vale para nada
mcompany_df = mcompany_df.drop(columns = ['unnamed:_0'])
```

```python
#Convertir 'effective_to_date' a tipo datetime
mcompany_df['effective_to_date'] = pd.to_datetime(mcompany_df['effective_to_date'])
```

```python
columns_list = mcompany_df.columns.tolist()

df_numerical_variables = mcompany_df.describe()

numerical_variables = df_numerical_variables.columns.tolist()

categorical_variables = list(filter(lambda x: x not in numerical_variables, columns_list))
```

```python
for i in categorical_variables:
    print(i)
    print(mcompany_df[i].unique())
```

```python
#Dado que la columna effective_to_date ya parece estar en formato datetime, no necesitas usar str.split(). En su lugar, puedes extraer directamente el año, mes y día utilizando los atributos .dt de pandas.

mcompany_df['year'] = mcompany_df['effective_to_date'].dt.year
mcompany_df['month2'] = mcompany_df['effective_to_date'].dt.month
mcompany_df['day'] = mcompany_df['effective_to_date'].dt.day
```

```python
#Comparamos month con month2

mcompany_df['month_comparison'] = mcompany_df['month'] == mcompany_df['month2']

#Solo me interesa las filas que no son iguales

non_matching_months = mcompany_df[mcompany_df['month'] != mcompany_df['month2']]
print(non_matching_months)

#Como no hay ninguna diferencia borro las columnas month_comparison y month2

mcompany_df = mcompany_df.drop(columns=['month_comparison', 'month2'])
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_df = mcompany_df.pivot_table(index='sales_channel', columns='month', values=['total_claim_amount'], aggfunc='sum')
pivot_df = pivot_df.round(2)
pivot_df
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Crear la tabla dinámica
pivot_df_lifetime = mcompany_df.pivot_table(
    index='gender',                      # Sexo como filas
    columns='education',                  # Nivel educativo como columnas
    values='customer_lifetime_value',     # Valor medio del ciclo de vida del cliente
    aggfunc='mean'                        # Función de agregación: media
)

# Redondear a 2 decimales para mejor presentación
pivot_df_lifetime = pivot_df_lifetime.round(2)

pivot_df_lifetime
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Crear la tabla pivote con el monto total de reclamaciones por tipo de póliza y mes
pivot_df_claims = mcompany_df.pivot_table(
    index='policy_type',               # Tipo de póliza en filas
    columns='month',                    # Mes en columnas
    values='total_claim_amount',        # Monto total de las reclamaciones
    aggfunc='sum'                       # Sumar el monto total de las reclamaciones
)

# Convertir la tabla a formato largo (pivotado)
pivot_long_df = pivot_df_claims.reset_index().melt(id_vars='policy_type', var_name='month', value_name='total_claim_amount')

# Asegurarse de que los meses sean tratados como números
pivot_long_df['month'] = pivot_long_df['month'].astype(int)

# Mostrar la tabla resultante en formato largo
pivot_long_df
```
